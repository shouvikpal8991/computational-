```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# Define the differential equation
def f(t, y):
    return t * np.exp(3*t) - 2 * y

# Define the time span
t_span = (0, 1)

# Define the initial condition
y0 = [0]

# Solve the ODE numerically
sol = solve_ivp(f, t_span, y0, t_eval=np.linspace(0, 1, 100))

# Define the analytical solution
def analytical_solution(t):
    return (1/25) * np.exp(-2* t) * (1 + (np.exp(5*t) *(-1+5*t)))

# Generate y-values for the analytical solution
t_values = np.linspace(0, 1, 100)
analytical_y = analytical_solution(t_values)

# Plotting the numerical and analytical solutions
plt.figure(figsize=(8, 6))
plt.plot(sol.t, sol.y[0], label='Numerical Solution')
plt.plot(t_values, analytical_y, label='Analytical Solution', linestyle='--')
plt.xlabel('t')
plt.ylabel('y(t)')
plt.title('Solution to y\' = te^(3t) - 2y, y(0) = 0')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# Define the differential equation
def f(t, y):
    return 1 - (t - y)**2

# Define the time span
t_span = (2, 3)

# Define the initial condition
y0 = [1]

# Solve the ODE numerically
sol = solve_ivp(f, t_span, y0, t_eval=np.linspace(2, 3, 100))

# Define the analytical solution
def analytical_solution(t):
    return (1 - 3 * t + t**2) / (-3 + t)

# Generate y-values for the analytical solution
t_values = np.linspace(2, 2.999, 100)
analytical_y = analytical_solution(t_values)

# Plotting the numerical solution
plt.figure(figsize=(8, 6))
plt.plot(sol.t, sol.y[0], label='Numerical Solution')
plt.xlabel('t')
plt.ylabel('y(t)')
plt.title('Numerical Solution to y\' = 1 - (t - y)^2, y(2) = 1')
plt.legend()
plt.grid(True)
plt.show()

# Plotting the analytical solution
plt.figure(figsize=(8, 6))
plt.plot(t_values, analytical_y, label='Analytical Solution', linestyle='--')
plt.xlabel('t')
plt.ylabel('y(t)')
plt.title('Analytical Solution to y\' = 1 - (t - y)^2, y(2) = 1')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# Define the differential equation
def f(t, y):
    return 1 + y / t

# Define the time span
t_span = (1, 2)

# Define the initial condition
y0 = [2]

# Solve the ODE numerically
sol = solve_ivp(f, t_span, y0, t_eval=np.linspace(1, 2, 100))

# Define the analytical solution
def analytical_solution(t):
    return t*(2+np.log(t))

# Generate y-values for the analytical solution
t_values = np.linspace(1, 2, 100)
analytical_y = analytical_solution(t_values)

# Plotting both solutions
plt.figure(figsize=(8, 6))

plt.plot(sol.t, sol.y[0], label='Numerical Solution')
plt.plot(t_values, analytical_y, label='Analytical Solution', linestyle='--')

plt.xlabel('t')
plt.ylabel('y(t)')
plt.title('Solutions to y\' = 1 + y/t, y(1) = 2')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# Define the differential equation
def f(t, y):
    return np.cos(2*t) + np.sin(3*t)

# Define the time span
t_span = (0, 1)

# Define the initial condition
y0 = [1]

# Solve the ODE numerically
sol = solve_ivp(f, t_span, y0, t_eval=np.linspace(0, 1, 100))

# Define the analytical solution
def analytical_solution(t):
    return (8 - 2 * np.cos(3 * t) + 3 * np.sin(2 * t)) / 6

# Generate y-values for the analytical solution
t_values = np.linspace(0, 1, 100)
analytical_y = analytical_solution(t_values)

# Plotting both solutions
plt.figure(figsize=(8, 6))

plt.plot(sol.t, sol.y[0], label='Numerical Solution')
plt.plot(t_values, analytical_y, label='Analytical Solution', linestyle='--')

plt.xlabel('t')
plt.ylabel('y(t)')
plt.title('Solutions to y\' = cos(2t) + sin(3t), y(0) = 1')
plt.legend()
plt.grid(True)
plt.show()
```
